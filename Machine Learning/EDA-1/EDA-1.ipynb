# Rules for Missing Value Imputation

- If NA values are between 1 to 5%, drop na (rows)
- If NA value are 5 to 40%, fillna(replace)
- If NA values are greater than 50% in a column, drop that entire column

```python
df.hist()
plt.tight_layout()
```

```python
sns.boxplot(x = df['Solar'])
```

```python
ozone_median = df['Ozone'].median()
```

```python
df['Ozone'].fillna(ozone_median, inplace = True)
```

```python
df['Solar'].fillna(df['Solar'].mean(), inplace = True)
```

```python
df['Weather'].mode()[0]

 # For categorical column
```

```python
df['Weather'].fillna(df['Weather'].mode()[0], inplace = True)
```
