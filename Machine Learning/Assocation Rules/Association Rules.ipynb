```python
# Importing the libraries

import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import association_rules,apriori
```

```python
# Loading the dataset

df = pd.read_csv("E:\ExcelR\Data Science\Datasets\Titanic.csv")
```

```python
df
```

```python
data = pd.get_dummies(df)
data
```

```python
# Building the apriori algorithm

scores = apriori(data, min_support=0.2, use_colnames=True)
```

```python
scores
```

```python
# Building the assocaition rules

association_rules(scores)
```

```python
for i in df['Products']:
    print(i.split(','))
```

```python
def txt_split(txt):
    return txt.split(',')
```

```python
df['Products'].apply(txt_split)
```

```python
# List Comprehension

data = [i.split(',') for i in df['Products']]
```

```python
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
encoded_df = te.fit_transform(data)
```

```python
te.columns_
```

```python
data = pd.DataFrame(encoded_df, columns=te.columns_)
data
```

```python
data.replace(True,1,inplace=True)
data.replace(False,0,inplace=True)
```

```python
data
```

```python
score = apriori(data,min_support=0.2,use_colnames=True)
```

```python
score
```

```python
association_rules(score)
```
